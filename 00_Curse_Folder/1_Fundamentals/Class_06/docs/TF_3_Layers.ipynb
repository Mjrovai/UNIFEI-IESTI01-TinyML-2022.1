**Import Libraries**

```python
%tensorflow_version 2.x
import tensorflow as tf
import numpy as np
import math
import matplotlib.pyplot as plt
```

**Using 3 Layers**

```python
my_layer_1 = tf.keras.layers.Dense(units=2, input_shape=[1])
my_layer_2 = tf.keras.layers.Dense(units=2, input_shape=[1])
my_layer_3 = tf.keras.layers.Dense(units=1)

model = tf.keras.Sequential([my_layer_1, my_layer_2,  my_layer_3])

model.compile(optimizer='sgd',loss='mean_squared_error')

xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float) 
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500, verbose=0)
```

```python
print("weights layer_1 size: ", len(my_layer_1.get_weights()))
print("weights layer_2 size: ", len(my_layer_2.get_weights()))
print("weights layer_3 size: ", len(my_layer_3.get_weights()))
```

Cada "layer weights" é uma lista com 2 itens (sendo cada item um array): ws é lista[0] e bs é lista[1]

```python
print("ws layer_1 shape: ", my_layer_1.get_weights()[0].shape)
print("ws layer_2 shape: ", my_layer_2.get_weights()[0].shape)
print("ws layer_3 shape: ", my_layer_3.get_weights()[0].shape)

print("bs layer_1 shape: ", my_layer_1.get_weights()[1].shape)
print("bs layer_2 shape: ", my_layer_2.get_weights()[1].shape)
print("bs layer_3 shape: ", my_layer_3.get_weights()[1].shape)
```

Calculate Y para um X = 10

```python
value = 10.0
```

Obter os pesos (ws e bs) por cada um dos neuronios dos layers

TF_3_layers.png

```python
layer1_w1=(my_layer_1.get_weights()[0][0][0])
layer1_w2=(my_layer_1.get_weights()[0][0][1])
layer1_b1=(my_layer_1.get_weights()[1][0])
layer1_b2=(my_layer_1.get_weights()[1][1])

layer2_w1=(my_layer_2.get_weights()[0][0][0])
layer2_w2=(my_layer_2.get_weights()[0][1][0])
layer2_w3=(my_layer_2.get_weights()[0][0][1])
layer2_w4=(my_layer_2.get_weights()[0][1][1])
layer2_b1=(my_layer_2.get_weights()[1][0])
layer2_b2=(my_layer_2.get_weights()[1][1])

layer3_w1=(my_layer_3.get_weights()[0][0])
layer3_w2=(my_layer_3.get_weights()[0][1])
layer3_b=(my_layer_3.get_weights()[1][0])
```

Calcular o valor de Y manualmante

```python
layer_1_out_1= (layer1_w1 * value)+ layer1_b1
layer_1_out_2 = (layer1_w2 * value)+ layer1_b2

layer_2_out_1= (layer2_w1 * layer_1_out_1)+(layer2_w2 * layer_1_out_2)+layer2_b1
layer_2_out_2= (layer2_w3 * layer_1_out_1)+(layer2_w4 * layer_1_out_2)+layer2_b2

neuron3_output = (layer3_w1 * layer_2_out_1)+ (layer3_w2*layer_2_out_2) + layer3_b

print(neuron3_output)
```
